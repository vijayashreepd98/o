```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

### reading data

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
# 
X=data['param1'].values
Y=data['param2'].values
```

### calculating coefficient


```python
#mean x and y
mean_x=np.mean(X)
mean_y=np.mean(Y)
```

```python
n=len(X)
```

```python
#using formula for finding b1 and b0
numer=0
denom=0
for i in  range(n):
    numer+=(X[1]-mean_x)*(Y[1]-mean_y)
    denom+=(X[1]-mean_x)**2
    b1=numer/denom
    b0=mean_y-(b1*mean_x)
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
```

```python
#printing coefficien
b1=numer/denom
print('coefficient')
print(b1,b0)

```

```python
#plotting values and regression line
max_x=np.max(X)+100
min_x=np.min(X)-100
```

```python
#calculating the line values in x and y
y=b0+b1*X
```

```python
#ploting line
plt.plot(X,y,color='#58b970',label='Regression line')
```

```python
#ploting scatter points
plt.plot(X,y,color='#58b970',label='Regression line')

plt.scatter(X,Y,c='#ef5423',label='scatter point')
```
